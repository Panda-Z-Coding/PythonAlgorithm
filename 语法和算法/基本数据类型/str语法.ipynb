# 转换类方法


```python
from dataclasses import replace


s = 'pandaco'
print(s.title())
print(s.swapcase())
s1 = 'lcz love hmh'
print(s1.replace('lcz','hmh'))
```

# 查找类函数




```python
s = '21321323'
print(s.count('2',0,4))
print(s.find('4',0,4))
print(s.startswith('2',0,0))
```
