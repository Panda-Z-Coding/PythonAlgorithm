# 集合

- 储存不重复数据
- 唯一性
- 储存不可变数据：数字、字符串、元组
- {e1,e2,...}
- 无序性

## 创建集合
- 创建：
  - {e1,e2,...}
  - {}是 **空字典！！** 不是空集合
  - set(x) 把x转化为集合
  - set() 创建一个 *空集合*,set()就是空集的意思
## 集合基本操作





```python
s = {1,2,3,4}
print(type(s))
print(s)
s1 = [1,2,3,4,3,2]
print(set(s1))
s3 = {}
print(type(s3))
print(set(s3))

key = ['a','b','c']
value = [123,234,345]

dic1 = dict(zip(key,value))
print(dic1.items())
print(dic1.keys())
print(dic1.values())
print(dic1['a'])

s4 = {2,3}
print("s4是否是s的子集:",s4.issubset(s))
```

## 集合基本操作
- S = {1,2,3} T = {2,3,3}
- 交集: S & T
- 并集: S | T
- 差集: S - T 、 T - S
- 对称差集: S ^ T
## 函数基本操作


```python
S = {1,2,3}
T = {2,3,3}
print(S & T)
print(S | T)
print(S - T) 
print(S ^ T)

#! a = list(map(float,input().split())) 把输入的数据分割转化成float类型列表
'''
jupyter不支持输入
'''
s = set()
print(s)
```
